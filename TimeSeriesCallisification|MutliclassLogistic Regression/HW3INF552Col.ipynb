# Confidence Interval : 

```python
import numpy as np
import scipy.stats

def mean_confidence_interval(data, confidence=0.90):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```

# Feature ranking with recursive feature elimination and cross-validation
RFECV performs RFE in a cross-validation loop to find the optimal number or the best number of features. Hereafter a recursive feature elimination applied on logistic regression with automatic tuning of the number of features are  selected with cross-validation. 
Approach followed : 
for each l = 1..........20 :
  a. prepare the dataset i.e. break timeseries features according to l 
  b. append columns for bending and non - bending 
  c. perfrom rfecv. 
  d. calculate the score on the training datset. 
now select the best l corresponding to highest score

```python

def takeaverageCols(res): # to give the min max mean of the ts 1,2,6 i.e generating time domain features for d(ii)
    sum = 0
    for res in res:
        sum = sum + res
    return sum/2

def calculateTimeSeriesFeaturesModel(perActivityTotalInstance,featuresPerActivityPerInstanceModel):
    timeseriesSetModel = ["avgrss12","varrss12","avgrss13","varrss13","avgrss23","varss23"]
    n = len(perActivityTotalInstance)
    index = 0
    while index<n:
        bs = perActivityTotalInstance[index]
        resultperInstance = []
        for k in range(0,6,1):
            mini = bs[timeseriesSetModel[k]].min()
            mini = takeaverage(mini)
            maxi = bs[timeseriesSetModel[k]].max()
            maxi = takeaverage(maxi)
            mean = bs[timeseriesSetModel[k]].mean()
            mean = takeaverageCols(mean)
            std =  bs[timeseriesSetModel[k]].std()
            std =  takeaverageCols(std)
            median =  bs[timeseriesSetModel[k]].median()
            median =  takeaverageCols(median)
            resultperInstance.append(mini)
            resultperInstance.append(maxi)
            resultperInstance.append(mean)
            resultperInstance.append(std)
            resultperInstance.append(median)
        featuresPerActivityPerInstanceModel.append(resultperInstance)
        index = index + 1
    return featuresPerActivityPerInstanceModel
    
```

 (ii.)classifier has to be tested using the features extracted from the test set. i.e min max mean std median
 
 

```python
featuresPerActivityPerInstance = []
def takeaverageCols(res): # to give the min max mean of the ts 1,2,6 i.e generating time domain features for d(ii)
    sum = 0
    for res in res:
        sum = sum + res
    return sum/2
def calculateTimeSeriesFeatures(perActivityTotalInstance):
    timeseriesSet = ["avgrss12", "varrss12", "avgrss13", "varrss13", "avgrss23", "varss23"]
    n = len(perActivityTotalInstance)
    index = 0
    while index<n:
        bs = perActivityTotalInstance[index]
        resultperInstance = []
        for k in range(0,6,1):
            mini = bs[timeseriesSet[k]].min()
            mini = takeaverage(mini)
            maxi = bs[timeseriesSet[k]].max()
            maxi = takeaverage(maxi)
            mean = bs[timeseriesSet[k]].mean()
            mean = takeaverageCols(mean)
            std =  bs[timeseriesSet[k]].std()
            std =  takeaverageCols(std)
            median =  bs[timeseriesSet[k]].median()
            median =  takeaverageCols(median)
            resultperInstance.append(mini)
            resultperInstance.append(maxi)
            resultperInstance.append(mean)
            resultperInstance.append(std)
            resultperInstance.append(median)
        featuresPerActivityPerInstance.append(resultperInstance)
        index = index + 1
```

```python
# Compute macro-average ROC curve and ROC area
from scipy import interp
# First aggregate all false positive rates
import numpy as np
from itertools import cycle
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
lw = 2
# Then interpolate all ROC curves at this points
mean_tpr = np.zeros_like(all_fpr)
for i in range(n_classes):
    mean_tpr += interp(all_fpr, fpr[i], tpr[i])

# Finally average it and compute AUC
mean_tpr /= n_classes

fpr["macro"] = all_fpr
tpr["macro"] = mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])
# Plot all ROC curves
plt.figure()
plt.plot(fpr["micro"], tpr["micro"],
         label='micro-average ROC curve (area = {0:0.2f})'
               ''.format(roc_auc["micro"]),
         color='deeppink', linestyle=':', linewidth=4)

plt.plot(fpr["macro"], tpr["macro"],
         label='macro-average ROC curve (area = {0:0.2f})'
               ''.format(roc_auc["macro"]),
         color='navy', linestyle=':', linewidth=4)

colors = cycle(['aqua', 'darkorange', 'cornflowerblue'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=lw,
             label='ROC curve of class {0} (area = {1:0.2f})'
             ''.format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--', lw=lw)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Some extension of Receiver operating characteristic to multi-class')
plt.legend(loc="lower right")
plt.show()
```
