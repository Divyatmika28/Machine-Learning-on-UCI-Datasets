# 3.b
As shown in the plot, the optimal separating hyperplane has to be between the observations (2,1) and (2,2), and between the observations (4,3) and (4,4). So it is a line that passes through the points (2,1.5) and (4,3.5) which equation is
X1−X2−0.5=0.

```python
import matplotlib.pyplot as plt
import itertools
import numpy as np
#import plotly.tools as tls
fig, ax = plt.subplots()
x1 = [3, 2, 4, 1, 2, 4, 4]
x2 = [4, 2, 4, 4, 1, 3, 1]
colors = ["red", "red", "red", "red", "blue", "blue", "blue"]
for i in range(0,7,1):
    color = colors[i]
    ax.scatter(x1[i], x2[i], color=color)

#X1−X2−0.5=0.
#x1=x2+0.5
ylist = np.array(x1)
xlist = np.array(x2)
X,Y = np.meshgrid(xlist, ylist) # Create 2-D grid xlist,ylist values
F = Y-X-0.5  #  'Circle Equation
plt.contour(Y, X, F, [0], colors = 'k', linestyles = 'solid')
plt.show()

```
